```python
# importing libraries - they cover data loading, cleaning, analysis, and visualization.
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

plt.style.use("ggplot")

```

```python
# Loading the dataset

df1 = pd.read_csv('../data/fear_greed_index.csv')
df2 = pd.read_csv('../data/historical_data.csv')

df1.head(), df2.head()
```

```python
df2.tail(10)
```

```python

df1.info()
df1.describe(include="all")
```

```python
# convert timestamp (seconds) to datetime
df1['timestamp'] = pd.to_datetime(df1['timestamp'], unit='s')
```

```python
df1.shape
print(df1.isna().sum())
```

```python

print('Unique classifications:', df1['classification'].unique())
print('Number of unique classifications:', df1['classification'].nunique())
print('Classification counts:\n', df1['classification'].value_counts())
```

```python
# convert date column
df1['date'] = pd.to_datetime(df1['date'])

```

```python


# map sentiment to numeric
sent_map = {
    "Extreme Fear": 0,
    "Fear": 1,
    "Neutral": 2,
    "Greed": 3,
    "Extreme Greed": 4
}
df1['sentiment_score'] = df1['classification'].map(sent_map)
```

```python
df1.head()
```

```python
# ----- Step 4: Drop useless columns -----
```

```python
print(df2.info())
print(df2.describe(include="all"))
```

```python
print(df2.isna().sum())
```

```python
# Timestamp IST to datetime
df2['Timestamp IST'] = pd.to_datetime(df2['Timestamp IST'], format='%d-%m-%Y %H:%M')

# Unix timestamp (might be in ms)
df2['timestamp_dt'] = pd.to_datetime(df2['Timestamp'], unit='ms')

# create trade date for merging
df2['trade_date'] = df2['Timestamp IST'].dt.date
df2['trade_date'] = pd.to_datetime(df2['trade_date'])

df2.head()
```

```python
df2 = df2.drop(columns=["Timestamp"], axis=1)
```

```python
def clean_num(x):
    if isinstance(x, str):
        x = x.replace(',', '')
    return pd.to_numeric(x, errors='coerce')

df2['Execution Price'] = clean_num(df2['Execution Price'])
df2['Size Tokens'] = clean_num(df2['Size Tokens'])
df2['Size USD'] = clean_num(df2['Size USD'])
# df2['Closed PnL'] = clean_num(df2['ClosedPnL'])
df2['Fee'] = clean_num(df2['Fee'])

df2.head()
```

```python
df2['Side'] = df2['Side'].astype(str).str.upper()
df2['Direction'] = df2['Direction'].astype(str).str.upper()

df2.drop_duplicates(inplace=True)
```

```python
df2.head()
```

```python
cols_to_drop = ["Transaction Hash","Order ID","Trade ID","timestamp_dt","trade_date"]
df2 = df2.drop(columns=cols_to_drop, axis=1)
df2.head()
```

```python
df2['Timestamp IST'] = pd.to_datetime(df2['Timestamp IST'], errors='coerce')
df2['trade_date'] = df2['Timestamp IST'].dt.date
df2.head()
```

```python
df1['date'] = pd.to_datetime(df1['date'], errors='coerce')
df1['sentiment_date'] = df1['date'].dt.date
df1.head()
```

```python
merged = df2.merge(
    df1[['sentiment_date','classification','sentiment_score']],
    left_on='trade_date',
    right_on='sentiment_date',
    how='left'
)
merged.head()
```

```python
merged.to_csv('../data/new_merged_data.csv', index=False)
```

```python
merged['is_win'] = merged['Closed PnL'] > 0
merged[['Closed PnL', 'is_win']].tail()
```

```python
# STEP 2  Start EDA
# Does market sentiment affect trader performance?
```

```python
# 1. Average PnL by sentiment
merged.groupby('classification')['Closed PnL'].mean()
merged.groupby('classification')['Closed PnL'].mean().plot(kind='bar', title='Average Closed PnL by Market Sentiment')
plt.ylabel('Average Closed PnL')
plt.show()
# This tells you if traders earn more during:
# Fear
# Greed
# Extreme Greed
# Extreme Fear
# Neutral
# Expected insight (industry):
# Traders lose more in extreme greed, and trade less in extreme fear.
```

2. Win Rate by sentiment
This shows if traders win more during certain market moods.

```python

merged.groupby('classification')['is_win'].mean()
merged.groupby('classification')['is_win'].mean().plot(kind='bar', title='Win Rate by Market Sentiment')
plt.ylabel('Win Rate')
plt.show()
# This shows if traders win more during certain market moods.
```

3. Trade Size vs sentiment
High trade size during Greed usually indicates risk-taking.

```python
merged.groupby('classification')['Size USD'].mean()
merged.groupby('classification')['Size USD'].mean().plot(kind='bar', title='Average Trade Size by Market Sentiment')
plt.ylabel('Average Trade Size (USD)')
plt.show()
# High trade size during Greed usually indicates risk-taking.
```

4. Side (Buy/Sell) behavior vs sentiment
Questions answered:

Do traders BUY more during fear?

Do they SELL more during extreme fear?

```python
merged.groupby(['classification','Side']).size().unstack()
merged.groupby(['classification','Side']).size().unstack().plot(kind='bar', stacked=True, title='Trade Counts by Sentiment and Side')
plt.ylabel('Number of Trades')
plt.show()
```

 5. Volume vs sentiment
-This shows how active traders are in different moods.

```python
merged.groupby('classification')['Size Tokens'].sum()
merged.groupby('classification')['Size Tokens'].sum().plot(kind='bar', title='Total Volume (Tokens) by Market Sentiment')
plt.ylabel('Total Volume (Tokens)')         
plt.show()
```
